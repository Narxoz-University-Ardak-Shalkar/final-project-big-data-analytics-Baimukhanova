2.   ,    (    ,   datetime, float, int  . .).

```python
# fillna(0)       0- .
# astype(int)      integer (int)   
data['QUANTITYORDERED'] = data['QUANTITYORDERED'].fillna(0).astype(int)
```

```python
# pd.to_datetime() 'ORDERDATE'   datetime  .
# format='%d/%m/%Y'    (%d),  (%m),   (%Y)
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'], format='%d/%m/%Y', errors='coerce')
```

3.      ,    .

```python
#   : float64 ( )  int64 ( ) 
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

#      IQR
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25) # Q1   25%-  .
        Q3 = df[column].quantile(0.75) # Q3   75%-  .
        IQR = Q3 - Q1 # IQR     
        lower_bound = Q1 - 1.5 * IQR #  
        upper_bound = Q3 + 1.5 * IQR #   
        #           
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

#      
data_cleaned = remove_outliers_iqr(data, numerical_columns)

#       
original_size = data.shape[0]
cleaned_size = data_cleaned.shape[0]

# 
{
    "Original Size": original_size,
    "Cleaned Size": cleaned_size,
    "Rows Removed": original_size - cleaned_size
}
```

```python
# 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


#   :
# select_dtypes(include=['object', 'category'])      .

categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print("   :", categorical_columns)

# 1.    :
# \n    ,        .
# value_counts()       .
for column in categorical_columns:
    print(f"\n    '{column}':")
    print(data[column].value_counts())

    #   
    plt.figure(figsize=(8, 4)) #   
    # sns.countplot()       
    # y=column    (y )    
    # data=data       (data)   .
    # palette='viridis'    
    sns.countplot(y=column, data=data, palette='viridis')
    plt.title(f' : {column}') #   
    plt.ylabel(column) # -   
    plt.xlabel('') #    (x )  .
    plt.show() #       .

# 2.       
#  'STATUS'  'SALES'      . 
if 'STATUS' in data.columns and 'SALES' in data.columns:
    # data.groupby('STATUS')    data   'STATUS'   . 
    # .mean()     'SALES'    
    # .sort_values() -   
    grouped_data = data.groupby('STATUS')['SALES'].mean().sort_values()
    print("\n   :")
    print(grouped_data)

    #   
    # kind='bar'      .
    # figsize=(12, 8)      :
    # color='orange'       .
    grouped_data.plot(kind='bar', figsize=(12, 8), color='orange')
    plt.title('   ')
    plt.xlabel('')
    plt.ylabel(' ')
    plt.show()
```

```python
# 1.   "  " (Total Cost)
#      .
data_cleaned['TOTALCOST'] = data_cleaned['QUANTITYORDERED'] * data_cleaned['PRICEEACH']

# 2.     :
# 'STATUS'      .
#        map()  .
data_cleaned['STATUS_ENCODED'] = data_cleaned['STATUS'].map({
    'Shipped': 1,
    'Pending': 0,
    'Cancelled': -1
})

# 3.      
#   " " (1,    > 50,  0)
# data_cleaned['QUANTITYORDERED'] > 50   False True
# .astype(int) True  False   1  0  .
data_cleaned['LARGE_ORDER'] = (data_cleaned['QUANTITYORDERED'] > 50).astype(int)

# 4.      (,     )
import numpy as np
#         (, 10- 1 000 000- ),    "".
# log(0) ,    1 ,        .
data_cleaned['LOG_SALES'] = np.log(data_cleaned['SALES'] + 1)  #  1,     0

# 5.   
#   "   " (Sales per Unit)
#    (    )   (    )   .
data_cleaned['SALES_PER_UNIT'] = data_cleaned['SALES'] / data_cleaned['QUANTITYORDERED']

#      
print(data_cleaned.head())
```

```python
#     
def calculate_discount(sales):
    if sales > 700:
        return sales * 0.1  # 10%    
    else:
        return sales * 0.05  # 5%    

#     'SALES'
#  apply    SALES,          calculate_discount
# DISCOUNT    .
data['DISCOUNT'] = data['SALES'].apply(calculate_discount)

print(data[['ORDERNUMBER', 'SALES', 'DISCOUNT']].head())
```

```python
# CUSTOMERNAME       
#   CUSTOMERNAME_UPPER,        .
# lambda   
data['CUSTOMERNAME_UPPER'] = data['CUSTOMERNAME'].apply(lambda x: x.upper())

print(data[['CUSTOMERNAME', 'CUSTOMERNAME_UPPER']].head())
```
