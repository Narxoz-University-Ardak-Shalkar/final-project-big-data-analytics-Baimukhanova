3. Определите и устраните выбросы в данных, предоставив объяснение вашего подхода.

```python
# Числовые столбцы алу: float64 (жүздік сан) және int64 (бүтін сан) 
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Функция для удаления выбросов по IQR
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25) # Q1 — деректердің 25%-ын қамтитын мән.
        Q3 = df[column].quantile(0.75) # Q3 — деректердің 75%-ын қамтитын мән.
        IQR = Q3 - Q1 # IQR — квартильдер арасындағы интерквартильдік ауқым
        lower_bound = Q1 - 1.5 * IQR # миннимум мән
        upper_bound = Q3 + 1.5 * IQR # максимум мән 
        # миннимум мәнге тең немесе үлкен және максимум мәнге тең немесе кіші
        df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df

# шыққан мәнді осында сақтап тұр 
data_cleaned = remove_outliers_iqr(data, numerical_columns)

# Деректер өлшемін салыстыру до и после 
original_size = data.shape[0]
cleaned_size = data_cleaned.shape[0]

# Результаты
{
    "Original Size": original_size,
    "Cleaned Size": cleaned_size,
    "Rows Removed": original_size - cleaned_size
}
```

2. Анализируйте категориальные переменные, подсчитывая частоты, используя groupby() для суммирования категорий.

```python
# Импорты
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Категориялық айнымалыларды анықтау:
# select_dtypes(include=['object', 'category']) — мәтіндік немесе категориялық типтер таңдайды.

categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print("Категориальные переменные в данных:", categorical_columns)

# 1. Категориялық айнымалылардың жиілігін есептеу:
# \n — жаңа жолды бастайды, яғни әрбір айнымалы үшін бөлек жолда ақпарат көрсетіледі.
# value_counts() әрбір категориялық мәннің қанша рет қайталанатынын көрсетеді.
for column in categorical_columns:
    print(f"\nЧастоты значений для переменной '{column}':")
    print(data[column].value_counts())

    # Построение графика частот
    plt.figure(figsize=(8, 4)) # графиктің өлшемдерін орнатады
    # sns.countplot() — категориялық айнымалылардың жиіліктерін бейнелейтін графикті салады
    # y=column параметрі вертикальды оське (y осіне) категориялық айнымалының атын орнатады
    # data=data — бұл параметр арқылы деректер жиынын (data) графикке беріп отырмыз.
    # palette='viridis' — графиктегі түстер палитрасы
    sns.countplot(y=column, data=data, palette='viridis')
    plt.title(f'Распределение категорий: {column}') # графиктің атауын орнатады
    plt.ylabel(column) # у-ке столбецтің атын жазад
    plt.xlabel('Частота') # графиктің горизонтальды осьіне (x осіне) атау орнатады.
    plt.show() # бұл функция графикті экранда көрсету үшін қолданылады.

# 2. Группировка данных по одной из категориальных переменных
#  'STATUS' және 'SALES' бағандарының деректер жиынында бар екендігін тексереді. 
if 'STATUS' in data.columns and 'SALES' in data.columns:
    # data.groupby('STATUS') — бұл әдіс data деректер жиынын 'STATUS' айнымалысына негізделіп топтайды. 
    # .mean() — әрбір топ үшін 'SALES' бағанының орташа мәнін есептейді
    # .sort_values() - өсу бойынша сортироватьетеді
    grouped_data = data.groupby('STATUS')['SALES'].mean().sort_values()
    print("\nСредние продажи по статусу:")
    print(grouped_data)

    # Визуализация среднего значения
    # kind='bar' параметрі графиктің түрін бағанды етіп белгілейді.
    # figsize=(12, 8) — бұл параметр графиктің өлшемін белгілейді:
    # color='orange' — барлық бағандардың түсін сарғылт етіп орнатады.
    grouped_data.plot(kind='bar', figsize=(12, 8), color='orange')
    plt.title('Средние продажи по статусу')
    plt.xlabel('Статус')
    plt.ylabel('Средние продажи')
    plt.show()
```

1. Используйте пользовательские функции с методами Pandas, такими как apply(), для управления данными.


```python
# Определим функцию для вычисления скидки
def calculate_discount(sales):
    if sales > 700:
        return sales * 0.1  # 10% скидка для больших продаж
    else:
        return sales * 0.05  # 5% скидка для меньших продаж

# Применим функцию к столбцу 'SALES'
# Метод apply применяется к столбцу SALES, и для каждого значения в этом столбце вызывается функция calculate_discount
# DISCOUNT деген жаңа столбец құрады.
data['DISCOUNT'] = data['SALES'].apply(calculate_discount)

print(data[['ORDERNUMBER', 'SALES', 'DISCOUNT']].head())
```

```python
# CUSTOMERNAME дегі мәндерді кіші әріптерді үлкен әріпке ауыстырады
# новый столбец CUSTOMERNAME_UPPER, где имена клиентов будут преобразованы в верхний регистр.
# lambda эта анонимная функция
data['CUSTOMERNAME_UPPER'] = data['CUSTOMERNAME'].apply(lambda x: x.upper())

print(data[['CUSTOMERNAME', 'CUSTOMERNAME_UPPER']].head())
```

```python
# pivot_table() — это метод Pandas, который создает сводную таблицу.
pivot_sales_by_country = data.pivot_table(
    values='SALES',           # Столбец для агрегации
    index='COUNTRY',          # Указывает, что таблица будет группироваться по столбцу COUNTRY (страна).
    aggfunc='sum',            # Функция агрегации (сумма)
    fill_value=0              # Заполнение пустых значений нулями
)

# Печать результатов
print(pivot_sales_by_country)
```
